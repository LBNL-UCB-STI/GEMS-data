```python
import osmnx
import pandas as pd
import geopandas as gpd
from multiprocessing import Pool
import time
import os
```

```python
path = 'C:/FHWA_R2/Network/CleanData/OSMNX/'
```

```python
# Read cleaned data
hpms_files = [file for file in os.listdir(path) if file.startswith("osmnx_stats_by_tract_chunk")]
files_df = pd.DataFrame({"filename": hpms_files})
obs = len(files_df)
print(obs)
files_df.head()
```

```python
osm_df = pd.read_csv(os.path.join(path, files_df.iloc[0]["filename"]))
for i in range(1,obs):
    file_path = os.path.join(path, files_df.iloc[i]["filename"])
    df = pd.read_csv(file_path)
    osm_df = pd.concat([osm_df, df], ignore_index=True)
```

```python
osm_df.head()
```

```python
print(osm_df.columns)
osm_df.describe()
```

```python
###################
# Compute metrics #
###################
# Intersection density [intersections/acre]
osm_df['intersection_density']= 4047*osm_df.intersection_count/osm_df.ALAND
# Self-loop proportion [0,1]
# Street density [miles/acre]
osm_df['street_density'] = 0.000621371*4047*osm_df['street_length_total']/osm_df.ALAND
# Average street length [miles]
osm_df['avg_street_length'] = 0.000621371*osm_df['street_length_avg']
# Average circuitry [0,inf]
# Average streets per node
# Node and edge count
osm_df.rename(columns={'n':'node_count','m':'edge_count'},inplace=True)
# Get useful variables
osm_df['street_count']=osm_df['streets_per_node_avg']*osm_df['node_count']
osm_df['edge_length_total']=osm_df['edge_length_total']*0.000621371
```

```python
osm_df.columns
```

```python
osm_df[['STATEFP','TRACTCE', 'GEOID','ALAND',
        'intersection_density','self_loop_proportion','street_density','avg_street_length','circuity_avg','streets_per_node_avg','node_count',
       'edge_count', 'street_count','edge_length_total']].to_csv('{}/osm_metrics.csv'.format(path))
```

```python
### Generate Paritioned Data ###
partitions = pd.read_csv('C:/FHWA_R2/Network/CleanData/final_partition_results.csv')
```

```python
partition_df = osm_df.merge(partitions,how='right',left_on='GEOID',right_on='GEOID')
print(partition_df.shape)
partition_df.drop(partition_df.loc[partition_df.unitedID != partition_df.unitedID].index,inplace=True)
partition_df.shape
```

```python
partition_df.columns
```

```python
#############################################
# Metric computation                        #
#############################################
partition_df['self_loop_x_edge_count'] = partition_df['self_loop_proportion']*partition_df['edge_count']
partition_df['edge_length_circuity'] = partition_df['edge_length_total']/partition_df['circuity_avg']
final_partition_df=partition_df[['unitedID','edge_count','street_count','intersection_count','node_count',
                                 'edge_length_total','street_length_total','ALAND',
                                 'self_loop_x_edge_count','edge_length_circuity']].groupby('unitedID').sum().reset_index()
final_partition_df["self_loop_proportion"] = final_partition_df['self_loop_x_edge_count']/final_partition_df['edge_count']
final_partition_df["street_density"] = 0.000621371*4047*final_partition_df['street_length_total']/final_partition_df['ALAND']
final_partition_df["avg_street_length"] = 0.000621371*final_partition_df['street_length_total']/final_partition_df['street_count']
final_partition_df["circuity_avg"] = final_partition_df['edge_length_total']/final_partition_df['edge_length_circuity']
final_partition_df["streets_per_node_avg"]=final_partition_df['street_count']/final_partition_df['node_count']
```

```python
final_partition_df[['unitedID','ALAND',
        'self_loop_proportion','street_density','avg_street_length','circuity_avg','streets_per_node_avg','node_count',
       'edge_count', 'street_count','edge_length_total']].to_csv('{}/osm_metrics_partition.csv'.format(path))
```
